```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('Advertising.csv')
df.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
sns.pairplot(df)
print(plt.scatter(x=df['Newspaper'], y=df['Sales']))
```

```python
# train test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(
    df.iloc[:, :-1], df.iloc[:, -1], test_size=0.2, random_state=42
)
```

```python
x_train.shape, x_test.shape, y_train.shape, y_test.shape
```

```python
# applying linear regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
lr.fit(x_train, y_train)
y_pred=lr.predict(x_test)
```

```python
y_pred
```

```python
# metrics r2_score
from sklearn.metrics import r2_score,mean_absolute_error
mean_absolute_error(y_test, y_pred)
```

```python
r2_score(y_test, y_pred)
```

```python
# Predicting sales
def predict_sales(tv, radio, newspaper, online_ad):
    features = np.array([[tv, radio, newspaper, online_ad]])
    result = lr.predict(features)
    return result[0]

```

```python
# contors on vedio
import cv2

# Open video capture (0 for webcam or use 'video.mp4' for file)
cap = cv2.VideoCapture(0)

while True:
    ret, frame = cap.read()
    if not ret:
        break

    # Convert to grayscale and apply threshold
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    ret, thresh = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)

    # Find contours
    contours, hierarchy = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    # Draw contours on original frame
    cv2.drawContours(frame, contours, -1, (0, 255, 0), 2)

    # Show the result
    cv2.imshow("Contours on Video", frame)

    # Press 'q' to exit
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import matplotlib.pyplot as plt

cap = cv2.VideoCapture(r"C:\Users\shubh\Downloads\3129576-uhd_3840_2160_30fps.mp4")

while True:
    ret, frame = cap.read()
    if not ret:
        break

    frame = cv2.resize(frame, (640, 480))
    gray = cv2.cvtColor(frame, cv2.COLOR_BGR2GRAY)
    blurred = cv2.GaussianBlur(gray, (5, 5), 0)
    ret, thresh = cv2.threshold(blurred, 60, 255, cv2.THRESH_BINARY)

    contours, _ = cv2.findContours(thresh, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)

    for contour in contours:
        area = cv2.contourArea(contour)
        if area > 400:
            approx = cv2.approxPolyDP(contour, 0.02 * cv2.arcLength(contour, True), True)
            if len(approx) == 3:
                shape = "Triangle"
            elif len(approx) == 4:
                shape = "Rectangle"
            elif len(approx) > 10:
                shape = "Circle"
            else:
                shape = "Other"

            cv2.drawContours(frame, [approx], -1, (0, 255, 0), 2)
            x, y = approx[0][0]
            cv2.putText(frame, shape, (x, y - 10),
                        cv2.FONT_HERSHEY_SIMPLEX, 0.6, (0, 255, 255), 2)

    # Convert BGR to RGB for display in matplotlib
    plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    break  # Remove this if you want to keep processing

cap.release()
```

```python
import cv2
import matplotlib.pyplot as plt

haar_cascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
cam = cv2.VideoCapture(0)

while True:
    ret, img = cam.read()
    if not ret:
        print("Failed to grab frame")
        break

    grayImg = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    faces = haar_cascade.detectMultiScale(grayImg, scaleFactor=1.3, minNeighbors=4)

    for (x, y, w, h) in faces:
        cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

    # Show in notebook instead of cv2.imshow()
    plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()

    break  # Remove this if you want continuous detection

cam.release()
```

```python
import cv2
import easyocr
import os
import matplotlib.pyplot as plt

# Avoid duplicate library error for some systems
os.environ['KMP_DUPLICATE_LIB_OK'] = 'True'

# Create EasyOCR reader
reader = easyocr.Reader(['en'])

# Image path
image_path = r"C:\Users\shubh\OneDrive\Desktop\nlp.py\ml\Screenshot 2025-08-04 133021.png"

# Read text from image
results = reader.readtext(image_path)

# Load the image
img = cv2.imread(image_path)

if results:  # Check if any text is found
    for bbox, text, prob in results:
        # bbox is [[top_left], [top_right], [bottom_right], [bottom_left]]
        top_left = tuple(map(int, bbox[0]))
        bottom_right = tuple(map(int, bbox[2]))

        # Draw rectangle & text
        cv2.rectangle(img, top_left, bottom_right, (0, 255, 0), 2)
        cv2.putText(img, text, (top_left[0], top_left[1] - 10),
                    cv2.FONT_HERSHEY_SIMPLEX, 0.5, (0, 255, 0), 2)

        print(f"Detected: '{text}' with confidence {prob:.2f}")
else:
    print("No text detected in the image.")

# Show the result
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()
```
